```python
import csv
import mysql.connector as connection

mydb = connection.connect(host= "localhost", database="dress_data", user= "root", passwd= "mysql", use_pure= True)
cur = mydb.cursor()

with open("Attribute Data.csv",'r')as f:
    next(f)
    data = csv.reader(f,delimiter="\n")    
    print(data)
    for i in data:
        s = i[0].split(',')
        print(s)
        a = ", ".join('"' + e + '"' for e in s)
        print(a)
        cur.execute(f"INSERT INTO attribute VALUES({a})")
    print("All the data is inserted")
    
mydb.commit()        
```

#### 6. in sql task try to perform left join operation with attrubute dataset and dress dataset on column Dress_ID

```python
import mysql.connector as connection

mydb = connection.connect(host= "localhost", database="dress_data", user= "root", passwd= "mysql", use_pure= True)
cur = mydb.cursor()
cur.execute('''select * from attribute 
left join dress 
on attribute.Dress_ID = dress.Dress_ID''')
cur.fetchall()
```
