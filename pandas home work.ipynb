#### 14. total percentage of active hour for a month 

```python
def percentage_calculation(time):
    time = str(time)
    h, m, s = time.split(':')
    sec = int(h) * 3600 + int(m) * 60 + int(s)
    time_per = (sec/86400)*100
    return time_per
```

```python
df_login["Total_Duration_Percentage"] = df_login['Duration'].apply(percentage_calculation)
```

```python
df_login['week_day'] = df_login['Convert_Date_Time'].apply(week_number_of_month)
```
