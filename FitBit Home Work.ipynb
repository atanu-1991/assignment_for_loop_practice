```python
header = ["Id","ActivityDate","TotalSteps","TotalDistance","TrackerDistance","LoggedActivitiesDistance","VeryActiveDistance","ModeratelyActiveDistance","LightActiveDistance","SedentaryActiveDistance","VeryActiveMinutes","FairlyActiveMinutes","LightlyActiveMinutes","SedentaryMinutes","Calories"]
csvfile = open("FitBit data.csv",'r')
reader = csv.DictReader(csvfile)

for each in reader:
    row={}
    for field in header:
        row[field]=each[field]
    collection.insert_one(row)
```
