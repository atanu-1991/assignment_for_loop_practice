```python
# q16:
def dict_element(**kwargs):
    l = []
    for i,j in kwargs.items():
        if type(j) == dict:
            for k,v in j.items():
                l.append(v)
                
        else:
            
            l.append(j)
        
    print(l)    


dict_element(a=1,b=2,c=2,d={"aa":1})
```

```python
# q21:
import os

def doc_file(files):
    doc_ext = [".doc",".docx"]
    doc_list = []
    for file in files:
        if os.path.splitext(file)[1].lower() in doc_ext:
            doc_list.append(file)
#             print(os.path.splitext(file)[0].lower())
            
            
    return doc_list


dir_files = os.listdir()
doc_file(dir_files)
        
```

```python
# q25:
import imaplib
import email
import traceback 

MAIL_ID = "atanukundu1991@gmail.com"

MAIL_PASSWORD = "Atanukunducse1052"

SMTP_SERVER = "imap.gmail.com"

SMTP_PORT = 993


def read_email_from_gmail():
    
    try:
        mail = imaplib.IMAP4_SSL(SMTP_SERVER,SMTP_PORT)
        mail.login(MAIL_ID,MAIL_PASSWORD)
        mail.select('inbox')
        
        data = mail.search(None, 'ALL')
        mail_ids = data[1]
        id_list = mail_ids[0].split()   
        first_email_id = int(id_list[0])
        latest_email_id = int(id_list[-1])
        
        
        for (index,i) in enumerate(range(latest_email_id,first_email_id,-1)):
            data = mail.fetch(str(i),'(RFC822)')
            for response_part in data:
                arr = response_part[0]
                if isinstance(arr,tuple):
                    msg = email.message_from_string(str(arr[1],'utf-8'))
                    
                    email_subject = msg['subject']
                    email_from = msg['from']
                    date = msg['Date']
                    
                    print("date: ",date,'\n')
                    print(index+1,'From: '+email_from+'\n')
                    print('Subject: '+str(email_subject)+'\n')
                    print("-"*100)
                    
                    
            if(index>5):
                break
                    
        
        
    except Exception as e:
        traceback.print_exc() 
        print(str(e))
        

read_email_from_gmail()
```
